### Docker file for the database I used found [here](https://github.com/ghusta/docker-postgres-world-db)

```python
import psycopg2 as pg
import pandas.io.sql as psql
```

```python
dbname = "world-db"
host = "localhost"
port = "5432"
user = "world"
password = "world123"

try:
    conn = pg.connect(database=dbname, host=host, port=port, user=user, password=password)
    df = psql.read_sql('SELECT * FROM city', conn)
finally:
    conn.close() 
```

```python
# Make sure the connection is closed
conn.closed
```
